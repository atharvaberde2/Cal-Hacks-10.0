```python
import re
import pandas as pd
import sklearn
import sklearn.neighbors
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.cluster import KMeans, MeanShift
import numpy as np
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

date = ""


file = open('tttt.html')
x = []
y = []

cols = ['Date', 'Time', 'Precipitation', 'Air temp (max)', 'Air temp (min)']
dtypes = {'Date': 'int', 'Time': 'int', 'Precipitation': 'float', 'Air temp (max)': 'int', "Air temp (min)": 'int'}

for line in file:
    if line.startswith('SANJOSE.C,'):
        y.append(line)
for line in y:
    data = {'Year': [int(line.split(',')[1][0:4]) for line in y],'Month' : [int(line.split(',')[1][4:6]) for line in y], 'Day' : [int(line.split(',')[1][6:8]) for line in y] , 'Time': [int(line.split(',')[2]) for line in y], 'Precipitation': [float(line.split(',')[3]) for line in y], 'Air temp (max)': [int(line.split(',')[5]) for line in y], 'Air temp (min)': [int(line.split(',')[6]) for line in y]}

df = pd.DataFrame(data)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
param_grid = {
    'n_estimators': [10,20,30,40,50, 100],  # Number of trees in the forest
    'max_depth': [None, 10, 20, 30],  # Maximum depth of the trees
    'min_samples_split': [2, 5, 10],  # Minimum samples required to split an internal node
    'min_samples_leaf': [1, 2, 4]  # Minimum samples required to be a leaf node
    #'n_neighbors' : [1,2,3,4,5,6,7]
}

X = df.drop(columns='Precipitation')
X = X.drop(columns='Air temp (max)')
X = X.drop(columns='Air temp (min)')

Y1 = df['Precipitation']
Y2 = df['Air temp (max)']
Y3 = df['Air temp (min)']

#######################################################################
X_train, X_test, Y1_train, Y1_test = train_test_split(X,Y1,test_size=0.33)

# Create the RandomForestRegressor model
model = RandomForestRegressor(random_state=42)
#model1 = KNeighborsRegressor(n_neighbors=17)
# Create a grid search object with cross-validation
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_absolute_error', n_jobs=-1)

# Fit the grid search to your training data
model.fit(X_train, Y1_train)

# Get the best model and hyperparameters

#best_params1 = grid_search.best_params_

# Make predictions on the test set with the best model
#model.score(X_test,Y1_test)
grid_search.fit(X_train, Y1_train)
mean_absolute_error(Y1_test, grid_search.predict(X_test))
```

```python
Y1_pred = model.predict(X_test)

# Calculate Mean Absolute Error (MAE) to evaluate the model
mae1 = mean_absolute_error(Y1_test, Y1_pred)
```

```python
X3_train, Y3_train, X_test, Y3_test = train_test_split(X,Y3,test_size=0.33, random_state=42)
X_train, X_test, Y2_train, Y2_test = train_test_split(X,Y2,test_size=0.33)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
Input=[('polynomial', PolynomialFeatures(include_bias=False,degree=6)),('ss',StandardScaler() ), ('model', Ridge(alpha = 0.1))]
pipe = Pipeline(Input)
```

```python
pipe.fit(X_train,Y2_train)
```

```python
pipe.score(X_test,Y2_test)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
kf = KFold(n_splits=10)
scores = cross_val_score(pipe, X, Y2, cv=kf)
print('Average validation score for min temperature:',np.mean(scores))
```

```python
X3_train, X3_test,Y3_train, Y3_test = train_test_split(X,Y3,test_size=0.33)
from sklearn.ensemble import GradientBoostingRegressor
gb = GradientBoostingRegressor(n_estimators=200, learning_rate=0.1, max_depth=5)
gb.fit(X3_train,Y3_train)
```

```python
gb.score(X3_test,Y3_test)
```

```python
from sklearn.ensemble import VotingRegressor
vr= VotingRegressor(estimators=[('gb', gb), ('rf', model), ('lr', pipe)])
```

```python
vr.fit(X3_train,Y3_train)
```

```python
vr.score(X3_test,Y3_test)
```

```python
kf = KFold(n_splits=10)
scores = cross_val_score(gb, X, Y3, cv=kf)
scores1 = cross_val_score(vr, X, Y3, cv=kf)
print('Average validation score for max temp through GradientBoostingRegressor:',np.mean(scores))
print('Average validation score for max temp through VotingRegressor:',np.mean(scores1))
```
